### 2D Array Operations

```python

l = [
    [1, 2, 3],
    [4, 69, 6],
    [7, 8, 9]
]
print(f"Our 2D Array:\n{l}")
element_69 = l[1][1]
print(f"The element at index [1][1] is: {element_69}")
another_element = l[0][2]
print(f"The element at index [0][2] is: {another_element}")
```

### More 2D Array Indexing Operations

```python

print(f"Current 2D Array:\n{l}")

first_row = l[0]
print(f"First row: {first_row}")

second_row = l[1]
print(f"Second row: {second_row}")

last_element_last_row = l[-1][-1]
print(f"Last element of the last row: {last_element_last_row}")

first_column = [row[0] for row in l]
print(f"First column: {first_column}")

second_column = [row[1] for row in l]
print(f"Second column: {second_column}")
```

### Accessing Each Element of a New 2D Array

```python

l = [
    [69, 2, 3],
    [4, 69, 6]
]

print(f"The new 2D Array:\n{l}")
print("\nAccessing each element:")

for row_index, row in enumerate(l):
    for col_index, element in enumerate(row):
        print(f"Element at [{row_index}][{col_index}]: {element}")
```

### Accessing Elements without Loops

To access each element individually without using loops, you can specify the exact row and column index for each element. For a 2D array like `l = [[1, 2, 3], [4, 69, 6]]`, the elements can be accessed as follows:

```python
print(f"The new 2D Array: {l}")

print(f"Element at [0][0]: {l[0][0]}")
print(f"Element at [0][1]: {l[0][1]}")
print(f"Element at [0][2]: {l[0][2]}")
print(f"Element at [1][0]: {l[1][0]}")
print(f"Element at [1][1]: {l[1][1]}")
```

### Conditional Concatenation Reversing

Let's consider two lists and a condition to select elements. We'll then concatenate the selected elements and reverse the resulting list.

```python

list1 = [69, 2, 3, 4, 5]
list2 = [69, 11, 12, 13, 14]

print(f"Original List 1: {list1}")
print(f"Original List 2: {list2}\n")


conditionally_selected_elements = []

for num in list1:
    if num % 2 == 0:
        conditionally_selected_elements.append(num)

for num in list2:
    if num % 2 != 0:
        conditionally_selected_elements.append(num)

print(f"Conditionally concatenated list: {conditionally_selected_elements}")

reversed_concatenated_list = conditionally_selected_elements[::-1]

print(f"Reversed conditionally concatenated list: {reversed_concatenated_list}")
```

### Array Slicing Operations

Slicing is a powerful feature in Python to extract a portion of a list. The general syntax for slicing is `[start:stop:step]`:
- `start`: The index where the slice begins (inclusive). If omitted, defaults to 0.
- `stop`: The index where the slice ends (exclusive). If omitted, defaults to the length of the list.
- `step`: The increment between elements (e.g., `2` for every other element). If omitted, defaults to 1.

```python
l = [
    [69, 2, 3],
    [4, 69, 6]
]

print(f"Original Array (l): {l}\n")


first_row_slice = l[0]
print(f"First row using slicing: {first_row_slice}")
second_row_slice = l[1]
print(f"Second row using slicing: {second_row_slice}\n")

sub_section_first_row = l[0][0:2]
print(f"Elements from index 0 to 1 of the first row: {sub_section_first_row}")

every_other_element = l[0][::2]
print(f"Every other element from the first row: {every_other_element}\n")

reversed_first_row = l[0][::-1]
print(f"Reversed first row: {reversed_first_row}")

reversed_array = l[::-1]
print(f"Reversed entire 2D array (row order): {reversed_array}")
```

### Array Extending and Splitting

Let's demonstrate how to extend a Python list (which serves as a basic array) by adding elements or another list, and then how to split a list into multiple parts.

```python

basic_array = [10, 20, 30, 40, 50]
print(f"Original Array: {basic_array}\n")

basic_array.append(60)
print(f"Array after appending 60: {basic_array}")

basic_array.extend([70, 80])
print(f"Array after extending with [70, 80]: {basic_array}\n")

part1 = basic_array[0:5]

part2 = basic_array[5:]
print(f"Array split into two parts:")
print(f"  Part 1: {part1}")
print(f"  Part 2: {part2}\n")

part_a = basic_array[0:3]
part_b = basic_array[3:6]
part_c = basic_array[6:]
print(f"Array split into three parts:")
print(f"  Part A: {part_a}")
print(f"  Part B: {part_b}")
print(f"  Part C: {part_c}")
```
