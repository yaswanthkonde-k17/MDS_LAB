```python
import numpy as np
import time

size = 1000000

# Python list operations
list1 = list(range(size))
list2 = list(range(size))

start_time_list = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
end_time_list = time.time()
python_list_time = end_time_list - start_time_list
print(f"Python List addition time: {python_list_time:.6f} seconds")

# NumPy array operations
np_array1 = np.arange(size)
np_array2 = np.arange(size)

start_time_numpy = time.time()
result_np_array = np_array1 + np_array2
end_time_numpy = time.time()
numpy_array_time = end_time_numpy - start_time_numpy
print(f"NumPy Array addition time: {numpy_array_time:.6f} seconds")
```

```python
import numpy as np
import time

size = 1000000

# Python list operations (Multiplication)
list1 = [i for i in range(size)]
list2 = [i * 2 for i in range(size)] # Using different values for variety

start_time_list_mul = time.time()
result_list_mul = [list1[i] * list2[i] for i in range(size)]
end_time_list_mul = time.time()
python_list_mul_time = end_time_list_mul - start_time_list_mul
print(f"Python List multiplication time: {python_list_mul_time:.6f} seconds")

# NumPy array operations (Multiplication)
np_array1_mul = np.arange(size)
np_array2_mul = np.arange(size) * 2

start_time_numpy_mul = time.time()
result_np_array_mul = np_array1_mul * np_array2_mul
end_time_numpy_mul = time.time()
numpy_array_mul_time = end_time_numpy_mul - start_time_numpy_mul
print(f"NumPy Array multiplication time: {numpy_array_mul_time:.6f} seconds")
```

```python
import sys
import numpy as np

size = 1000000  # Number of elements

# Memory usage for Python list
python_list = list(range(size))
# An approximation: get size of the list object itself + sum of sizes of its elements (integers)
# sys.getsizeof() gives the size of the list structure itself, not its contents in memory.
# Each integer in Python is an object, so we estimate its size and multiply by 'size'.
# A Python int typically takes 28 bytes for smaller values, but can vary.
memory_list = sys.getsizeof(python_list) + sum(sys.getsizeof(item) for item in python_list[:1000]) * (size // 1000) # Sample sum for larger lists for estimation
print(f"Memory used by Python list (approx): {memory_list / (1024 * 1024):.3f} MB")

# Memory usage for NumPy array
numpy_array = np.arange(size)
# NumPy arrays store data more efficiently. sys.getsizeof gives the actual memory.
memory_numpy = numpy_array.nbytes
print(f"Memory used by NumPy array: {memory_numpy / (1024 * 1024):.3f} MB")
```
