## Convert String to List (Array)

```python
# Method 1: Convert string to a list of characters
my_string = "hello"
char_list = list(my_string)
print(f"Original string: '{my_string}'")
print(f"List of characters: {char_list}")
```

```python
# Method 2: Split string by a delimiter into a list of substrings
sentence = "This is a sample sentence"
words = sentence.split(' ')
print(f"Original sentence: '{sentence}'")
print(f"List of words: {words}")

# Example with a different delimiter
comma_separated_values = "apple,banana,orange"
items = comma_separated_values.split(',')
print(f"Original values: '{comma_separated_values}'")
print(f"List of items: {items}")
```
